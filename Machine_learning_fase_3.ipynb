### **B. ¿Cuántos clúster sería adecuado seleccionar para representar los datos?**

#### - Selección del número óptimo de clústeres (k)Para determinar el número óptimo de grupos (k) necesarios para el análisis mediante el algoritmo K-Means, se evaluaron dos técnicas estándar:

#### El método del codo (Elbow Method) y el método de la silueta (Silhouette Method).

#### Tras la ejecución, se optó por el método del codo, cuyos resultados confirman que el valor óptimo es $k=2$.

#### Por consiguiente, se procederá a segmentar la información en dos clústeres diferenciados, lo cual permitirá un análisis eficiente y robusto bajo el modelo de K-Means.

```python
import matplotlib.pyplot as grafica_francisco_Elborn_Metod
from sklearn.cluster import KMeans
k_values = range(1,11)
wcss = []
for k in k_values:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(dataframe_1_entrenamiento)
    wcss.append(kmeans.inertia_)

grafica_francisco_Elborn_Metod.figure(figsize=(10, 6))
grafica_francisco_Elborn_Metod.plot(k_values, wcss, marker='o', linestyle='--')
grafica_francisco_Elborn_Metod.title('Método del Codo para tu Dataset')
grafica_francisco_Elborn_Metod.xlabel('Número de Clusters (k)')
grafica_francisco_Elborn_Metod.ylabel('WCSS (Inercia)')
grafica_francisco_Elborn_Metod.grid(True)
grafica_francisco_Elborn_Metod.show()
grafica_francisco_Elborn_Metod.savefig('grafica_francisco_Elborn_Metod.png')
```

#### - Metodo del silhouette_score


```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

silhouette_scores = []

K = range(2, 11)  

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(dataframe_1_entrenamiento)
    score = silhouette_score(dataframe_1_entrenamiento, labels)
    silhouette_scores.append(score)
    print(f"k={k}, silhouette={score:.3f}")
```

#### - Segun el metodo de silhouette_score deberia ser k=2 "k=2 -  silhouette=0.568" siendo k=2 el coeficiente mas alto

```python
# Utilizamos el método del gráfico de distancia k para ayudar a elegir un valor épsilon adecuado
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors

def plot_k_distance_graph(X, k):
    neigh = NearestNeighbors(n_neighbors=k)
    neigh.fit(X)
    distances, _ = neigh.kneighbors(X)
    distances = np.sort(distances[:, k-1])
    plt.figure(figsize=(10, 6))
    plt.plot(distances)
    plt.xlabel('Points')
    plt.ylabel(f'{k}-th nearest neighbor distance')
    plt.title('K-distance Graph')
    plt.show()
plot_k_distance_graph(pca_enfermedades, k=4)

### Aclaro el k de la funcion osea (n_neighbors=k), esta definido por el numero de columnas o variables y se usa la regla 2 x variables, en este caso al ser 2 columnas osea 2 variables seia k= 4 , para que nos brinde el codo con el numero de épsilon
```

```python
from sklearn.cluster import DBSCAN

eps_values = np.arange(0.01, 0.10, 0.01)
silhouette_scores = []
valid_eps_values = []

for eps in eps_values:
    db = DBSCAN(eps=eps, min_samples=5).fit(pca_enfermedades)
    labels = db.labels_
    if len(set(labels)) > 1 and len(set(labels) - {-1}) > 1:
        score = silhouette_score(pca_enfermedades, labels)
        silhouette_scores.append(score)
        valid_eps_values.append(eps)
if silhouette_scores:
    best_index = np.argmax(silhouette_scores)
    best_eps = valid_eps_values[best_index]
    max_score = silhouette_scores[best_index]
    
    print(f"El mejor valor de eps es: {best_eps:.2f} con un Silhouette Score de: {max_score:.4f}")
    plt.figure(figsize=(10, 5))
    plt.plot(valid_eps_values, silhouette_scores, marker='o', linestyle='-', color='b')
    plt.axvline(x=best_eps, color='r', linestyle='--', label=f'Mejor eps: {best_eps:.2f}')
    
    plt.title("Optimización de eps para DBSCAN usando Silhouette Score")
    plt.xlabel("Valor de eps (Epsilon)")
    plt.ylabel("Silhouette Score")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```
